```python
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
import re
import bs4
```

```python
def getClassname(number_of_dec):
    if number_of_dec == 0:
        return '<tr class="RowClass2">'
    elif number_of_dec == 1:
        return '<tr class="RowClass3">'
    elif number_of_dec == 2:
        return '<tr class="RowClass4">'
    elif number_of_dec == 3:
        return '<tr class="RowClass5">'
    else:
        return '<tr class="">'
```

```python
sub_set_df_fhtml = sub_set_df_fhtml.replace(np.nan, '', regex=True)
```

```python
html = sub_set_df_fhtml.to_html(index=False,header=False)
html
```

```python
text_file = open("index.html", "w")
text_file.write(html)
text_file.close()
```

```python
soup = BeautifulSoup(html,"html.parser")
tags = soup.find_all('tr',recursive=True)
```

```python
df[0][0]
```

```python
for num , tag in enumerate(tags):
#     if (num > 1):
        print(num)
        row_class = getClassname(str(df[0][num]).count('.'))
        tags[num] = str(tag).replace("<tr>", row_class)
        print(tag)
        tags[num] = BeautifulSoup(tags[num], 'lxml')
        tags[num] = tags[num].find_all('tr',recursive=True)[0]
```

```python
for tr in soup.find_all("tr"): 
    tr.decompose()
```

```python
for tag in tags:
    soup.tbody.append(tag)
```
